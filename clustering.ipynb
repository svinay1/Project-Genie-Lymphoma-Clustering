```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in and preprocess dummy variable matrix
dummy_variables = pd.read_csv('Data/dummy_variables.csv')
dummy_variables = dummy_variables.drop(['Unnamed: 0'], axis=1)
dummy_variables = dummy_variables.drop(['Standardized'], axis=1)
dv_copy = dummy_variables
dummy_variables = dummy_variables.drop(['Cancer Type Detailed'], axis=1)
```

```python
# Calculate top 21 mutations
sums = {}
muts = list(dummy_variables.head(1))[1:]
tot_sum = 0
for mut in muts:
    tot_sum += dummy_variables[mut].sum()
    if dummy_variables[mut].sum() < 740:
        dummy_variables.drop([mut], axis=1, inplace=True)
```

```python
# Calculate average
def f(s):
    return s/s.sum()
```

```python
# Drop necessary data
temp = dummy_variables.drop(["Sample ID"], axis = 1)
temp.apply(f)
```

```python
# Calculate frequency of unique mutations
unique = dummy_variables.nunique()
unique
```

```python
# Drop mutations with only one occurrence
mutations = list(dummy_variables.head(1))[1:]
for col in mutations:
    if unique[col] == 1:
        dummy_variables = dummy_variables.drop([col], axis=1)
```

```python
# Generate Gower's matrix
import gower

dist = gower.gower_matrix(dv_copy)
df = pd.DataFrame(dist)
```

```python
# Convert dataframe to an array
df = df.div(df.sum(axis=1), axis=0)
df = df.to_numpy()
```

```python
# Match each sample ID to a number from 0 to 20
lst = list(dummy_variables["Sample ID"])
length = len(list(dummy_variables["Sample ID"]))
samples = range(length)
zipped = list(zip(lst, samples))
print(zipped)
```

```python
# Get cancer type of each sample:
sample_to_cancer = {}
for item in zipped:
    sample_to_cancer[item[1]] = dv_copy.loc[dv_copy['Sample ID'] == item[0], 'Cancer Type Detailed'].values[0]
for key, val in sample_to_cancer.items():
    print(key, val)
```

```python
# Generate dendrogram
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import squareform

import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))
linkage_matrix = linkage(df, "ward")
dendrogram(linkage_matrix, labels=samples)
plt.show()
```
