```python
def standardize(sample_ids):
    standard_sample_ids = []
    for sample in sample_ids:
        if sample[6:11] == "MSK-P":
            standard = sample[:6] + "MSKP" + sample[11:]
        elif sample[6:9] == "UHN":
            standard = sample[:6] + "UHNA" + sample[9:]
        else:
            standard = sample
        standard_sample_ids.append(standard[11:])
    return standard_sample_ids
```

```python
# Method to calculate frequency
def frequency(dataframe, column):
    dictionary = {}
    for item in dataframe[column]:
        if item in dictionary:
            dictionary[item] += 1
        else:
            dictionary[item] = 1
    return dictionary
```

```python
# Function that maps each standardized sample ID to its mutation
def mapping(dataframe, column1, column2):
    dictionary = {}
    col1 = dataframe[column1]
    col2 = dataframe[column2]
    values = []
    for idx in range(len(col1)):
        dictionary[col1[idx]] = []
    for idx in range(len(col1)):
        dictionary[col1[idx]].append(col2[idx])
    return dictionary
```

```python
pattern = mapping(hugo_filtered, "Standardized", "Hugo_Symbol")
print(pattern)
```

```python
# Assign dummy variables as 0.0 temporarily
genes = set(hugo_filtered["Hugo_Symbol"])
idx = 2
for gene in genes:
    #patients.insert(loc = idx, column = gene, value = [0.0])
    patients[gene] = 0.0
    idx += 1
```

```python
# Change dummy variable to +1.0 if mutation and sample ID match
def create_dummy(df):
    sum = 0
    for row in hugo_filtered["Standardized"]:
        if row in pattern:
            for gene in pattern[row]:
                idx = df.index[df['Standardized'] == row]
                df.loc[idx, [gene]] += [1.0]
                sum += 1
    return df, sum
```

```python
(patients, tot) = create_dummy(patients)
print(tot)
```
